# Train with shifted input

```python
tf.reset_default_graph()
time_step=100
state_size=1
lstm_cell = tf.nn.rnn_cell.BasicLSTMCell(state_size, forget_bias=1.)
def_x = np.sin(np.linspace(0, 10, time_step))[None, :, None]
X = tf.placeholder_with_default(def_x, [None, time_step, 1])
pred, last_states = tf.nn.dynamic_rnn(inputs=X, cell=lstm_cell, dtype=tf.float64)


Y = np.roll(def_x, 1)
loss = tf.reduce_sum(tf.pow(pred-Y, 2))/(2*time_step)

plt.plot(def_x.squeeze())
plt.plot(Y.squeeze())

opt = tf.train.AdamOptimizer().minimize(loss)
sess = tf.InteractiveSession()
tf.global_variables_initializer().run()

for i in range(1000):
    p, l, _= sess.run([pred, loss, opt])
    if i%100 == 0: 
        #plt.plot(def_x.squeeze())
        #plt.show(plt.plot(p.squeeze()))
        print(i, l)
```

# How to evaluate model

## Mainly involves [SO Tuple solution](http://stackoverflow.com/questions/39112622/how-do-i-set-tensorflow-rnn-state-when-state-is-tuple-true)

```python
lstm_cell.state_size
```

```python
GEN_LEN = 200
with tf.variable_scope('sine', reuse=True):
    X = tf.placeholder(dtype=tf.float64, shape=[1, 1, 1])
    # second dim is batch size (i.e. =1)
    state = tf.placeholder(dtype=tf.float64, shape=[2, 1, state_size])
    l = tf.unstack(state, axis=0)
    state_tuple = tf.contrib.rnn.LSTMStateTuple(l[0], l[1])
    
    lstm_cell = tf.contrib.rnn.BasicLSTMCell(state_size, forget_bias=1.)
    
    output, last_states = tf.nn.dynamic_rnn(
        inputs=X, cell=lstm_cell, dtype=tf.float64, initial_state=state_tuple)
    
    pred = tf.contrib.layers.fully_connected(last_states[1], 1, activation_fn=tf.tanh)
    
    state_val = sess.run(state_tuple, {state:np.zeros((2, 1, state_size))})
    pred_val = [[[0.0]]]
    # Saving the state instead of using 'Unrolled LSTM' - because no backprop is needed
    feed_dict = {state_tuple:state_val, X:pred_val}
    
    outputs_mod20 = []
    for i in range(GEN_LEN):
        state_val, pred_val = sess.run([last_states, pred], feed_dict)
        # feed_dict = {state_tuple:state_val, X:[pred_val]}
        # TRY DIFFERENT MOD VALUES
        feed_dict = {state_tuple:state_val, X:[[def_x[0, i%20]]]}
        # print(i, pred_val)
        outputs_mod20.append(pred_val.squeeze())
```
